```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Media e soma dos valores da lista
n = [5,6,2,8,6,7]
print(np.mean(n))
print(np.sum(n))
```

```python
# Funções
def media(app):
    lista = [5,6,2,8,6,7]
    lista.append(app)
    m = np.mean(lista)
    return m
```

```python
med = media(3)
print(med)
```

```python
# List Comprehension
# Selecione apenas os valores divisores por 2 e retorne a media
n = np.random.randint(1, 100, size=10)
np.mean([x for x in n if x % 2 == 0]) # Usando compreensão para obter a media dos valores divididos por 2

```

```python
# Lista de valores bagunçados
nomes = ['joao  ', ' pedro', ' marcus', 'maria', ' lucia', 'ana']
```

```python
# Usando as funções para ritirar os espaços e deixando com a primeira letra maiuscula
[l.strip().capitalize() for l in nomes]
```

```python
# Exenplo de função que retorna ao quadrado e a media
def quad(valor):
    return np.mean(valor ** 2)
        
print(quad(n))
```

```python
# Expressões Lambda - retorna o quadrado e a media
lamb = lambda x : x ** 2 # função multiplica todos os valores da lista por 2
print(np.mean(lamb(n))) # passando o valor para a função retornando uma lista e a media
```

```python
# Expressões Lambda - retorna impar ou par
imp_par = lambda x : print('Par') if x % 2 == 0 else print('Impar')

imp_par(548843)
```

```python
# Loc -Localiza os dados baseado em uma condição
dados[['Sexo', 'Idade','Classe', 'Tarifa', 'Sobreviveu']]\
.loc[(dados['Idade'] >= 1) & (dados['Idade'] < 10) & (dados['Sobreviveu'] == 1)]
```

```python
dados[['Sexo', 'Idade','Classe', 'Tarifa', 'Sobreviveu']]\
.loc[(dados['Idade'] >= 14) & (dados['Sobreviveu'] == 1) & (dados['Sexo'] == 'female')]
```

```python
# Map - Mapea os valores e substitui pelo valor desejado
dados['Sexo'].map({'male':'homem', 'female':'mulher'})
```

```python
# Replace - Seleciona os valores e substitui pelo valor desejado, usando o inplace valor muda na base
dados['Sexo'].replace(['male','female'],['homem','mulher']) # adicionando inplace muda em base
```

```python
# Alternativa para mudar os valores sem precisar do inplace
dados['Sexo'] = dados['Sexo'].map({'male':'homem', 'female':'mulher'})
```

```python
dados['Sexo']
```

```python
# Groupby - agrupamento por coluna, usando sempre uma função
dados.groupby(['Sexo','Classe']).mean()
```

```python
# Ordenação - sort_values pode receber uma coluna como referência
dados[['Sexo', 'Idade','Classe', 'Tarifa', 'Sobreviveu']].sort_values('Classe', ascending=False)
```

```python
# Ordenando index - Sort_index
dados[['Sexo', 'Idade','Classe', 'Tarifa', 'Sobreviveu']].sort_index(ascending=False)
```

```python
# Dados faltando - NaN - Not a Number- função isnull para revelar os valores 
dados.isnull().sum()
```

```python
# Substituindo os valores nulos da coluna embarque
dados.Embarque.fillna('S', inplace=True)
```

```python
# Conferindo
dados.Embarque.isnull().sum()
```

```python
# Deletando coluna com valores faltando  - Drop
dados.drop('Cabine', axis=1, inplace=True)
dados.head()
```

```python
# Apply e Lambda - aplicando um função usando lambda - removendo ultimo digito dos bilhetes
dados['Bilhete'] = dados['Bilhete'].apply(lambda x: x[:-1])
dados['Bilhete']  # Analisando os numeros dos bilhetes para ver a familiariedade entre eles 
```

```python
# Feature Engineering - Engenharia de recurso
# Gerando coluna 0 ou 1 para Embarque atraves do Dummy
embarque = pd.get_dummies(dados['Embarque'], drop_first=True)
```

```python
# Variáveis Dummy | HotEncoder - gerando coluna 0 ou 1 para o sexo
dados = pd.get_dummies(dados, columns=['Sexo'], drop_first=True)
```

```python
# Visualizando
embarque
```

```python
# Concatenando as colunas
dados = pd.concat([dados, embarque], axis=1)
dados.drop(['Embarque'], axis=1, inplace=True)
```

```python
dados.head()
```

```python
# Expressões Regulares
dados['Titulo'] = dados['Nome'].str.extract('([a-zA-Z]+)\.')
```

```python
dados
```

```python
# Feature Engineering - Engenharia de recursos
# Retornando homens e mulheres para analise
dados['Sexo'] = dados['Sexo_mulher'].map({0: 'homem', 1: 'mulher'})
```

```python
# Retorne todos os valores da lista, caso seja diferente atribua com o valor Outros
dados['Titulo'] = dados['Titulo'].apply(lambda x: 'Outros' if x not in ['Master', 'Miss', 'Mr', 'Ms'] else x)
```

```python
# Visualizandos os titulos por variavel
pd.crosstab(dados['Titulo'], dados['Sexo'])
```
